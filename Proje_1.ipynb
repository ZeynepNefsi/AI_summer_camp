```python
#sınıfların okunması
path = '/content/PetImages'
class_name = os.listdir(path)
class_name
```

```python
#kedi ve köpek sınıflarını listeliyoruz ve eleman sayısını yazdırdık.
cats = os.listdir(path+class_name[0])
dogs = os.listdir(path+class_name[1])
len(cats),len(dogs)
```

```python
#kedi için görüntüleri okuyun, resizing (aynı boyuta getirme) işlemi yaptıktan sonra normalizasyon yapın.
cat_label = 0;
data=[]; non_read=[]
for i in cats:
    cat_img_path = os.path.join(path,class_name[0],i)
    try:
      img = cv.imread(cat_img_path)
      img = cv.resize(img,(128,128))
      img = ((img-img.min())/(img.max()-img.min())).astype('float32')
      data.append((img,cat_label))
    except:
      non_read.append(cat_img_path)
len(data),data[0]
```

```python
#aynı işlemler köpek için de yapılır.
dog_label = 1
for i in dogs:
    dog_img_path = os.path.join(path,class_name[1],i)
    try:
      img = cv.imread(dog_img_path)
      img = cv.resize(img,(128,128))
      img = ((img-img.min())/(img.max()-img.min())).astype('float32')
      data.append((img,dog_label))
    except:
      non_read.append(dog_img_path)
len(data),len(non_read)
```

```python
X = []
y = []

for image, idx in data:
  X.append(image)
  y.append(idx)
```

```python
X_train = []
y_train = []

X_val = []
y_val = []

X_test = []
y_test = []
```

```python
X_train = X[:1600]
y_train = y[:1600]

X_val = X[1600:1800]
y_val = y[1600:1800]

X_test = X[1800:]
y_test = y[1800:]
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)

X_val = np.array(X_val)
y_val = np.array(y_val)

X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
# bu dosyaları bilgisayarımda kaydedip başka sayfada bunları kullanıp model eğitimi yapacağız.
import pickle
with open('/content/drive/MyDrive/Colabnot/Transfer_learning/train.pickle','wb') as f:
    pickle.dump(X_train,f)
with open('/content/drive/MyDrive/Colabnot/Transfer_learning/test.pickle','wb') as f:
    pickle.dump(X_test,f)
with open('/content/drive/MyDrive/Colabnot/Transfer_learning/val.pickle','wb') as f:
    pickle.dump(X_val,f)
```

```python
import tensorflow as tf
base_model = tf.keras.applications.mobilenet_v2.MobileNetV2(include_top = False,
                                                     weights = 'imagenet',
                                                     input_shape = (128,128,3),
                                                     pooling="max")
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
x = tf.keras.layers.Flatten()(base_model.output)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
predictions = tf.keras.layers.Dense(1, activation = 'sigmoid')(x)
```

```python
head_model = tf.keras.Model(inputs = base_model.input, outputs = predictions)
```

```python
head_model.compile(optimizer='adam', loss=tf.keras.losses.binary_crossentropy, metrics=['accuracy'])
```

```python
history = head_model.fit(X_train, y_train, batch_size=128, epochs=25, validation_data=(X_val, y_val))
```
